# Algoritmo de regresi√≥n


```
#w1 = precio o valor
def hipotesis(w0,w1, crimen):
  return w0 + w1*crimen
```

```
alfa = 0.01
```

```
def calcular_w0(w0, w1, muestrasCrimen, muestrasPrecio):
  sumatoria = 0
  indiceMuestra = 0
  m = len(muestrasCrimen)
  for muestra in muestrasCrimen:
    h = hipotesis(w0, w1, muestra)
    y = muestrasPrecio[indiceMuestra]
    sumatoria += h - y
    indiceMuestra += 1
  sumatoria = alfa*(sumatoria/m)
  return w0-sumatoria
```

```
def calcular_w1(w1, w0, muestrasCrimen, muestrasPrecio):
  sumatoria = 0
  indiceMuestra = 0
  m = len(muestrasCrimen)
  for muestra in muestrasCrimen:
    h = hipotesis(w0, w1, muestra)
    y = muestrasPrecio[indiceMuestra]
    sumatoria += (h - y)* muestra
    indiceMuestra += 1
  sumatoria = alfa*(sumatoria/m)
  return w1-sumatoria
```

```
import random 

def calculoHipotesis(valor, w0, w1):
  hipos = []
  for v in valor:
    hipos.append(hipotesis(w0, w1, v))
  return  hipos

def regresion(crimen, valorTierra, iteraciones=10):
  referenciaX = crimen
  w0 = random.uniform(-10,10)
  w1 = random.uniform(-10,10)
  while(iteraciones):
    iteraciones = iteraciones - 1
    w0_temporal = calcular_w0(w0, w1, crimen, valores)
    w1_temporal = calcular_w1(w1, w0, crimen, valores)
    w0 = w0_temporal
    w1 = w1_temporal

    plt.scatter(crimen, valores, color="blue")
    plt.plot(referenciaX, calculoHipotesis(crimen, w0, w1), color="red")
    plt.show()
```

```
regresion(crimen, valores, iteraciones = 500)
```
